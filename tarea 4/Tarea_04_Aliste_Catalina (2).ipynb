```python
import matplotlib.pyplot as plt
eje_x = ["lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"]
datos_1 = [4, 5, 5, 6, 7, 9, 4]
datos_2 = [6, 4, 8, 5, 5, 2, 5]
datos_3 = [5, 5, 7, 4, 6, 9, 5]
datos_4 = [6, 7, 7, 5, 5, 9, 6]

plt.plot(eje_x, datos_1, label = "Semana 1", color = "green")
plt.plot(eje_x, datos_2, label = "Semana 2", color = "pink")
plt.plot(eje_x, datos_3, label = "Semana 3", color = "violet")
plt.plot(eje_x, datos_4, label = "Semana 4", color = "blue")

plt.title("Horas de sueño de Cata del mes de septiembre 2023")
plt.xlabel("Días de la semana")
plt.ylabel("Horas")

plt.legend()
plt.show()

#Seleccioné la lobrería de Matplot porque era la que conocía
#Elegí hacer un gráfico de líneas porque me pareció que mostraría mejor los cambios drásticos de horas de sueño
#La historia que se puede ver al analizar el gráfico es el de una persona (yo, Catalina) con no muy buenos hpabitos de sueño, cuya vida diaria podría estar implicada en esto
#Podemos sacar conlcusiones a partir de estos datos como el que el miércole sy sábado son los días donde mas duerme, por lo tanto son los que no trabaja o asiste a la Universidad
#También podemos concluir que de la semana 2, el día que menos durmió fue el sábado, quizás por qué...
#Por otro lado, sus horas varían lo suficiente coo para pasar d eun extremo a otro, como de 4 (poco saludable) a 9 (tampoco muy saludable) tomando en cuenta que 7/8 hora sson las recomendables
```
